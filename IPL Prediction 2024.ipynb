```python
# Filter the data for the years 2021 to 2023
ipl_data = df[(df['season'] >= 2020) & (df['season'] <= 2023)]

# Display the filtered data
ipl_data.head(), ipl_data.shape
```

```python
# Implementing the missing value handling strategy

# Filling missing categorical data with 'Unknown'
categorical_columns = ['toss_won', 'decision', 'home_captain', 'away_captain', 'pom', 'super_over',
                       'highlights', 'home_key_batsman', 'home_key_bowler', 'home_playx1', 'away_playx1',
                       'away_key_batsman', 'away_key_bowler', 'match_days', 'umpire1', 'umpire2',
                       'tv_umpire', 'referee', 'reserve_umpire']
ipl_data[categorical_columns] = ipl_data[categorical_columns].fillna('Unknown')

# Filling missing numerical data with 0 (assuming absence of records means the event did not occur)
numerical_columns = ['home_overs', 'home_runs', 'home_wickets', 'home_boundaries',
                     'away_overs', 'away_runs', 'away_wickets', 'away_boundaries']
ipl_data[numerical_columns] = ipl_data[numerical_columns].fillna(0)

# For 'season', 'points', and 'winner' where filling with 'Unknown' or a specific strategy might be needed
ipl_data['season'] = ipl_data['season'].fillna(ipl_data['season'].mode()[0])  # Most common season
ipl_data['points'] = ipl_data['points'].fillna('Unknown')  # Points might need specific handling
ipl_data['winner'] = ipl_data['winner'].fillna('No Result')  # For matches without a winner

# Check if any null values remain
ipl_data.isnull().sum().sum()
# Normalize Data Formats
# Convert dates to datetime format, ensuring consistency.
ipl_data['start_date'] = pd.to_datetime(ipl_data['start_date'])
ipl_data['end_date'] = pd.to_datetime(ipl_data['end_date'])

```

```python
# Re-calculating 'average_home_runs' and ensuring it's added to 'ipl_data'
# For simplicity, let's assume home team always bats first, which may not always be the case
ipl_data['1st_inning_score'] = pd.to_numeric(ipl_data['1st_inning_score'], errors='coerce')
ipl_data['average_home_runs'] = ipl_data.groupby('home_team')['1st_inning_score'].transform('mean')

# Confirm the column is now in 'ipl_data'
'average_home_runs' in ipl_data.columns
# Re-defining the 'home_team_win' column
ipl_data['home_team_win'] = (ipl_data['home_team'] == ipl_data['winner']).astype(int)

# Confirming the 'home_team_win' column is now part of the DataFrame
'home_team_win' in ipl_data.columns

                                              
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder

# Simplify 'toss_won' to a binary feature where 1 indicates the home team won the toss, and 0 otherwise
ipl_data['toss_won_by_home_team'] = (ipl_data['home_team'] == ipl_data['toss_won']).astype(int)

# Selecting a subset of features for modeling
features = ipl_data[['average_home_runs', 'toss_won_by_home_team']]
target = ipl_data['home_team_win']

# Handling missing values in the features (replacing NaN with the median for 'average_home_runs')
features['average_home_runs'] = features['average_home_runs'].fillna(features['average_home_runs'].median())

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

### Step 3: Train the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)

# Predicting outcomes on the test set
y_pred = rf_classifier.predict(X_test)

### Step 4: Evaluate the Model
# Calculating accuracy and displaying the classification report
accuracy = accuracy_score(y_test, y_pred,"\n")
classification_rep = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("\nClassification Report:\n", classification_rep)
```
